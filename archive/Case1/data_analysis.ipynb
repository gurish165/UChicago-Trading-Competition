```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def data_analysis(year):
    pricedf = pd.read_csv(f'prices_{year}.csv', index_col=0)
    raindf = pd.read_csv(f'rain_{year}.csv', index_col=0)
    fig, axs = plt.subplots(2, figsize=(10,10))
    axs[0].plot(pricedf)
    for i in range(12):
        axs[0].axvline(x=(i+1)*21, color='g', linestyle='--')
    axs[0].set_title(f'Price of Lumber for each trading day on {year}')
    axs[0].set_xlabel('Trading day')
    axs[0].set_ylabel('Price')

    axs[1].bar(raindf.index+1, raindf['Historical Monthly Rain'])
    axs[1].set_title(f'Precipitation of Lumber for each trading month {year}')
    axs[1].set_xlabel('Month')
    axs[1].set_ylabel('Precipitation')
    plt.show()
```

```python
data_analysis(2016)
```

```python
data_analysis(2017)
```

```python
data_analysis(2018)
```

```python
data_analysis(2019)
```

```python
data_analysis(2020)
```
